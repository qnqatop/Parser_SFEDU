импортируем нужные библиотеки

```python
import requests #библиотека для запросов
from bs4 import BeautifulSoup   #библиотека для парсинга
from time import sleep  #стандартная библиотека для задержки между запросами
import pandas as pd  #библиотека для работы с данными
```

делем запрос через **requests.get**

в качестве примера взял сайт со страницей магазина айфонов 13 мини

```python
url = 'http://apple-rostov.com/katalog/apple-iphone/apple-iphone-13-mini/'
r = requests.get(url)
r.text
```

отображем в более понятном виде **html** кода

```python
soup = BeautifulSoup(r.text, 'lxml')
soup
```

работа непосредственно с кодом страницы
находим имя 

```python
name = soup.find('div', class_="item_product col-md-3 col-xs-6").find('div',class_='imit').find('img').get('alt')
name
```

находим цену

```python
price = soup.find('div', class_="item_product col-md-3 col-xs-6").find('div',class_='price').text
price
```

находим ссылку

```python
link = soup.find('div', class_="item_product col-md-3 col-xs-6").find('a',class_='iteml').get('href')
link
```

прогоняем через цикл и прохоим по все айфонам на старнице
записваем в виде массива айфон по типу имя,цена,ссылка

```python
data = []

iphones = soup.findAll('div', class_="item_product col-md-3 col-xs-6")

for iphone in iphones:
  name = iphone.find('div',class_='imit').find('img').get('alt')
  price = iphone.find('div',class_='price').text + 'р'
  link = iphone.find('a',class_='iteml').get('href')
  data.append([name,price,link])



```

отображаем и смотрим что все отработало

```python
data
```

преобразуем в виде адекватной таблицы с названивиями столбцов в виде полей которые находили

```python
haders = ['name','price','link']
date = pd.DataFrame(data, columns=haders)
```

проверяем

```python
date
```

в моем случае я сохранил в виде csv на свой гугл диск

```python
date.to_csv('/content/drive/MyDrive/apple.csv',sep=';',encoding='utf8')
```

отсортируем по цена-название

```python
new_df = date.groupby(['price','name']).mean()
new_df
```
